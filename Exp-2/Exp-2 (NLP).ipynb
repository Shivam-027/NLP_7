```python
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter
import string
nltk.download("punkt")
```

```python
def find_frequent_words(text, n=5):
    tokens = word_tokenize(text.lower())

    word_freq = Counter(tokens)

    most_common_words = word_freq.most_common(n)

    return most_common_words

text = "Generative AI (GenAI) refers to a subset of artificial intelligence technologies designed to generate content, such as text, images, audio, and even video, based on data patterns. Unlike traditional AI, which primarily focuses on analyzing and interpreting data, GenAI creates new content that mimics human creativity. It leverages advanced models like deep learning and neural networks, such as transformers, to understand input and generate realistic or creative outputs. Common applications include natural language processing for text generation, image synthesis from sketches or descriptions, music composition, and even generating conversational agents. GenAI has revolutionized fields like art, design, entertainment, and content creation, allowing users to automate and augment creative processes."

top_words = find_frequent_words(text)
print(top_words)
```

```python
def find_frequent_words(text, n=5):
    tokens = word_tokenize(text.lower())

    tokens = [word for word in tokens if word not in string.punctuation]

    word_freq = Counter(tokens)

    most_common_words = word_freq.most_common(n)

    return most_common_words

text = "Generative AI (GenAI) refers to a subset of artificial intelligence technologies designed to generate content, such as text, images, audio, and even video, based on data patterns. Unlike traditional AI, which primarily focuses on analyzing and interpreting data, GenAI creates new content that mimics human creativity. It leverages advanced models like deep learning and neural networks, such as transformers, to understand input and generate realistic or creative outputs. Common applications include natural language processing for text generation, image synthesis from sketches or descriptions, music composition, and even generating conversational agents. GenAI has revolutionized fields like art, design, entertainment, and content creation, allowing users to automate and augment creative processes."

top_words = find_frequent_words(text)
print(top_words)
```
