# Detect Ususual Words

```python
common_words = set(["list", "of", "common", "words", "in", "our", "language"])
```

```python
def find_unusual_words(text, common_words):
    words = text.split()
    unusual_words = set()
    for word in words:
        word = word.lower()
        if word not in common_words:
            unusual_words.add(word)
    return unusual_words
```

```python
text = "This is an example text in your language with some unusual words."
unusual_words = find_unusual_words(text, common_words)
```

```python
print("Unusual words in the text:")
for word in unusual_words:
    print(word)
```
