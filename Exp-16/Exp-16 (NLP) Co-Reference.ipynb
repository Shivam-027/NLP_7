```python
import os
from huggingface_hub import login
from fastcoref import FCoref

hf_token = "API_TOKEN"
login(hf_token)

coref = FCoref()
```

```python
def coreference_resolution(text):
    result = coref.predict(texts=[text])

    resolved_text = result[0]

    return resolved_text
```

```python
text = "John is a programmer. He works at Google."
new_text = coreference_resolution(text)
print(f"Text: {text}")
print(f"New Text: {new_text}")
```
