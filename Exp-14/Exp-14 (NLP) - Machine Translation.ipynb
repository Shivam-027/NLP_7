## **Using Model**

```python
from transformers import MarianMTModel, MarianTokenizer

def translate_text(text, src_language="es", target_language="en"):
    model_name = f"Helsinki-NLP/opus-mt-{src_language}-{target_language}"
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)

    tokenized_text = tokenizer([text], return_tensors="pt", padding=True)
    translated = model.generate(**tokenized_text)

    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)
    return translated_text
```

```python
text_to_translate = "Hola, ¿cómo estás?"
translated_text = translate_text(text_to_translate, src_language="es", target_language="en")
print(f"Translated Text: {translated_text}")
```
