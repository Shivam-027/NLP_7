```python
import docx
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
# Load the document
doc = docx.Document('/content/english-en.docx')

# Get all the sentences in the document
sentences = []
for paragraph in doc.paragraphs:
    sentences.append(paragraph.text)

# Create a PorterStemmer object
stemmer = PorterStemmer()

# Create a WordNetLemmatizer object
lemmatizer = WordNetLemmatizer()
```

```python
# Stemming
stemmed_sentences = []
for sentence in sentences:
    stemmed_sentence = []
    for word in sentence.split():
        stemmed_word = stemmer.stem(word)
        stemmed_sentence.append(stemmed_word)
    stemmed_sentences.append(' '.join(stemmed_sentence))
```

```python
# Print stemmed sentences
print("Stemmed Sentences:")
# for s in stemmed_sentences:
#     print(s)
print(stemmed_sentences)
```

```python
# Lemmatization
lemmatized_sentences = []
for sentence in sentences:
    lemmatized_sentence = []
    for word in word_tokenize(sentence):
        lemmatized_word = lemmatizer.lemmatize(word)
        lemmatized_sentence.append(lemmatized_word)
    lemmatized_sentences.append(' '.join(lemmatized_sentence))
```

```python
print("Lemmatized Sentences:")
# for l in lemmatized_sentences:
#     print(l)
print(lemmatized_sentences)
```

# Stopword Removal

```python
import docx
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Load the document
doc = docx.Document("/content/english-en.docx")

# Get all the sentences in the document
sentences = []
for paragraph in doc.paragraphs:
    sentences.append(paragraph.text)

# List of stopwords
stop_words = set(stopwords.words('english'))

# Remove stopwords from all the sentences
stopword_removed_sentences = []
for sentence in sentences:
    stopword_removed_sentence = []
    for word in word_tokenize(sentence):
        if word.lower() not in stop_words:
            stopword_removed_sentence.append(word)
    stopword_removed_sentences.append(' '.join(stopword_removed_sentence))
```

```python
# Print the sentences with stopwords removed
print("Sentences with stopwords removed:")
# for s in stopword_removed_sentences:
#     print(s)
print(stopword_removed_sentences)
```
