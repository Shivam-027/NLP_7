```python
import re
from nltk.tokenize import word_tokenize
import nltk
from nltk.corpus import sentiwordnet as sns

# Ensure NLTK tokenizer data is downloaded
nltk.download('punkt')
nltk.download('sentiwordnet')
```

```python
# Define preprocess function
def preprocess(text):
    try:
        # Convert text to lowercase
        text = text.lower()

        # Remove punctuation and special characters using regex
        text = re.sub(r'[^\w\s]', '', text)

        # Tokenize the text into words
        words = word_tokenize(text)

        return words

    except Exception as e:
        print(f"Error in preprocessing the text: {text}")
        print(f"Exception: {e}")
        return []

# Define the calculate_sentiment_score function
def calculate_sentiment_score(words):
    pos_score = 0
    neg_score = 0

    for word in words:
        synsets = list(sns.senti_synsets(word))
        if synsets:
            pos_score += synsets[0].pos_score()
            neg_score += synsets[0].neg_score()

    # Return both positive and negative scores
    return pos_score, neg_score

# Define analyze_sentiment function
def analyze_sentiment(text):
    words = preprocess(text)
    if not words:
        return "Error in preprocessing"

    pos_score, neg_score = calculate_sentiment_score(words)

    if pos_score > neg_score:
        return 'Positive'
    elif neg_score > pos_score:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
if __name__ == "__main__":
    hotel_reviews = [
        "The hotel service was excellent, but the room was not clean.",
        "Amazing experience, Friendly staff, and beautiful rooms.",
        "Disappointing stay. Terrible service and dirty rooms.",
        "I would not recommend this hotel to anyone. Bad services."
    ]

    for i, review in enumerate(hotel_reviews, start=1):
        sentiment = analyze_sentiment(review)
        print(f"Review {i}:")
        print(f"Text: {review}")
        print(f"Sentiment: {sentiment}")
        print("\n")
```
